# Transformation

```python
#import dependencies
import pandas as pd
import json
import requests
import numpy as np
```

```python
Rental_df = pd.read_csv("Average Rents.csv")
Centres_df = pd.read_csv("Centres.csv")
Zones_df = pd.read_csv("Zones.csv")
Neighbourhood_df = pd.read_csv('Neighbourhood.csv')
Dwelling_df = pd.read_csv('Dwelling.csv')
Province_df =pd.read_csv("Provinces.csv")
```

```python
Dwelling_df.keys()
```

```python
Rental_df_copy = Rental_df.copy()
```

```python
Rental_df_copy.keys()
```

```python
Rental_df_copy = Rental_df_copy.rename(columns = { "Centre" : "Centres", "Zone" : "Zones" ,'Dwelling Type ' : 'Dwelling Type','Province': 'Provinces'})
Rental_df_copy.keys()
```

```python
Rental_cleaned = pd.merge(Rental_df_copy,Province_df, on = 'Provinces' , how = 'left')
Rental_cleaned = pd.merge(Rental_cleaned,Centres_df, on = 'Centres' , how = 'left')
Rental_cleaned = pd.merge(Rental_cleaned,Zones_df, on = 'Zones' , how = 'left')
Rental_cleaned = pd.merge(Rental_cleaned,Neighbourhood_df, on = 'Neighbourhood' , how = 'left')
Rental_cleaned = pd.merge(Rental_cleaned,Dwelling_df, on = 'Dwelling Type' , how = 'left')
Rental_cleaned 
```

```python
Rental_cleaned.keys()
```

```python
length_df = (len(Rental_cleaned))
```

```python
# Create unique ids 
Rent_ID = np.arange(1, length_df + 1)
Rental_id = ['Rent_' + str(each) for each in Rent_ID]
#print(Rental_id)
```

```python
Rental_cleaned["Rent ID"] = Rental_id
```

```python
# check data types
Rental_cleaned.dtypes
```

```python
Rental_cleaned = Rental_cleaned.rename(columns = { 'Bachelor ': 'Bachelor', 'Total ' : 'Total'})
Rental_cleaned.keys()
```

```python
Rental_cleaned = Rental_cleaned[['Rent ID','Bachelor', '1 Bedroom', '2 Bedroom', '3 Bedroom +', 'Total', 'Year','Centre_id', 'Zone_id', 'Neighbourhood_id', 'Dwelling_id' ]]
```

```python
Rental_cleaned
```

```python
Rental_cleaned.to_csv("Rental Data.csv", index=False)
```

```python
Vacancy_df = pd.read_csv('Vacancy Rates.csv')
```

```python
Vacancy_df_copy = Vacancy_df.rename(columns = { "Centre" : "Centres", "Zone" : "Zones",'Province': 'Provinces'})
Vacancy_df_copy.keys()
```

```python
Vacancy_df_copy = Vacancy_df_copy.rename(columns = { 'Bachelor ': 'Bachelor', 'Total ' : 'Total', 'Dwelling Type ' : 'Dwelling Type'})
Vacancy_df_copy
```

```python
Vacancy_cleaned = pd.merge(Vacancy_df_copy,Province_df, on = 'Provinces' , how = 'left')
Vacancy_cleaned = pd.merge(Vacancy_cleaned,Centres_df, on = 'Centres' , how = 'left')
Vacancy_cleaned = pd.merge(Vacancy_cleaned,Zones_df, on = 'Zones' , how = 'left')
Vacancy_cleaned  = pd.merge(Vacancy_cleaned ,Neighbourhood_df, on = 'Neighbourhood' , how = 'left')
Vacancy_cleaned  = pd.merge(Vacancy_cleaned ,Dwelling_df, on = 'Dwelling Type' , how = 'left')
Vacancy_cleaned 
```

```python
# Vacancy_cleaned = Vacancy_merged_df.drop(['Centres','Zones'], axis = 1)
# Vacancy_cleaned.head()
Vacancy_ID = np.arange(1, length_df + 1)
Vacancy_id = ['VR_' + str(each) for each in Vacancy_ID]
Vacancy_cleaned["VR_ID"] = Vacancy_id
Vacancy_cleaned
```

```python
Vacancy_cleaned = Vacancy_cleaned[['VR_ID','Bachelor', '1 Bedroom', '2 Bedroom', '3 Bedroom +', 'Total', 'Year','Centre_id', 'Zone_id', 'Neighbourhood_id', 'Dwelling_id' ]]
```

```python
Vacancy_cleaned.to_csv("Vacancy Rate Data.csv", index=False)
```

```python
Units_df = pd.read_csv('Number of Units.csv')
Units_df_copy = Units_df.rename(columns = { "Centre" : "Centres", "Zone" : "Zones", 'Dwelling Type ' : 'Dwelling Type','Province': 'Provinces'})
Units_df_copy = Units_df_copy.rename(columns = { 'Bachelor ': 'Bachelor', 'Total ' : 'Total'})
Units_cleaned = pd.merge(Units_df_copy,Province_df, on = 'Provinces' , how = 'left')
Units_cleaned = pd.merge(Units_cleaned,Centres_df, on = 'Centres' , how = 'left')
Units_cleaned = pd.merge(Units_cleaned,Zones_df, on = 'Zones' , how = 'left')
Units_cleaned  = pd.merge(Units_cleaned ,Neighbourhood_df, on = 'Neighbourhood' , how = 'left')
Units_cleaned = pd.merge(Units_cleaned ,Dwelling_df, on = 'Dwelling Type' , how = 'left')

Units_ID = np.arange(1, length_df + 1)
Units_id = ['Units_' + str(each) for each in Units_ID]
Units_cleaned["Units_ID"] = Units_id

Units_cleaned

#Units_cleaned = Units_merged_df.drop(['Centres','Zones'], axis = 1)
Units_cleaned.head()
```

```python
Units_cleaned = Units_cleaned[['Units_ID','Bachelor', '1 Bedroom', '2 Bedroom', '3 Bedroom +', 'Total', 'Year','Centre_id', 'Zone_id', 'Neighbourhood_id', 'Dwelling_id' ]]
```

```python
Units_cleaned.to_csv("Number of Units Data.csv", index=False)
```
