```python
#read initial files
rental_df = pd.read_csv("Average Rents.csv")
units_df = pd.read_csv("Number of Units.csv")
vacancy_df = pd.read_csv("Vacancy Rates.csv")
```

```python
Centre_ids_number = np.arange(1,len(Centres)+1 )
print(len(Centre_ids_number))
Centre_ids = ['centre' + str(each) for each in Centre_ids_number]
print(Centre_ids)
```

```python
Centre_df = pd.DataFrame({"Centre_id" : Centre_ids , 'Centres' : Centres})
Centre_df
```

```python
Zones =  rental_df["Zone"].unique()
print(Zones)
```

```python
Zones_ids_number = np.arange(1,len(Zones)+1 )
Zone_ids = ['Zone' + str(each) for each in Zones_ids_number]
#print(Centre_ids)
Zone_df = pd.DataFrame({"Zone_id" : Zone_ids , 'Zones' : Zones})
Zone_df
```
